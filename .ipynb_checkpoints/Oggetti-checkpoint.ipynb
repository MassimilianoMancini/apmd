````python
class Point:
    def __init__(self, initX, initY):
        """Create a new point at the given coordinates"""
        self.x = initX
        self.y = initY
        
    def getX(self):
        return self.x
    
    def getY(self):
        return self.y
    
    def distanceFromOrigin(self):
        return ((self.x ** 2) + (self.y ** 2)) ** 0.5
    
    def __str__(self):
        return f'({self.x}, {self.y})'
    
    def halfway(self, target):
        mx = (self.x + target.x) / 2
        my = (self.y + target.y) / 2
        return Point(mx, my)
    
    def distanceFromPoint(self, target):
        return (((self.x - target.x) ** 2) + ((self.y - target.y) ** 2)) ** 0.5
    
    def reflect_x(self):
        return Point(self.x * -1, self.y)
````

L'equazione di una linea retta è $$y = ax + b$$ oppure $$y = mx + c$$ I coefficienti $a$ e $b$ descrivono completamente la linea. Scrivere un metodo nella classe `Point` in modo che, dato un altro punto come parametro, calcola l'equazione della linea passante per il punto (`self`) e il punto dato. Deve ritornare i due coefficienti come tuple. Per esempio
```
>>> print(Point(4, 11).get_line_to(Point(6, 15)))
>>> (2, 3)
```
Questo ci dice che l'equazione della linea che passa per i due punti è $$y = 2x + 3$$. Quando fallirà il metodo?

````python
class Point:
    def __init__(self, initX, initY):
        """Create a new point at the given coordinates"""
        self.x = initX
        self.y = initY
        
    def getX(self):
        return self.x
    
    def getY(self):
        return self.y
    
    def distanceFromOrigin(self):
        return ((self.x ** 2) + (self.y ** 2)) ** 0.5
    
    def __str__(self):
        return f'({self.x}, {self.y})'
    
    def halfway(self, target):
        mx = (self.x + target.x) / 2
        my = (self.y + target.y) / 2
        return Point(mx, my)
    
    def distanceFromPoint(self, target):
        return (((self.x - target.x) ** 2) + ((self.y - target.y) ** 2)) ** 0.5
    
    def reflect_x(self):
        return Point(self.x * -1, self.y)
    
    def slopeFromOrigin(self):
        return self.y/self.x if self.x != 0 else None
    
    def get_line_to(self, target):
        d = self.x - target.x
        if d == 0:
            return None
        m = (self.y - target.y)/d
        b = self.y - m * self.x
        return (m, b)
````

Aggiungere un metodo move che prende due parametri, chiamati dx e dy. Il metodo farà muovere il punto nella direzione x e y rispettivamente di un numero di unità pari a dx e dy (i.e. cambiare lo stato del punto)

````python
class Point:
    def __init__(self, initX, initY):
        """Create a new point at the given coordinates"""
        self.x = initX
        self.y = initY
        
    def getX(self):
        return self.x
    
    def getY(self):
        return self.y
    
    def distanceFromOrigin(self):
        return ((self.x ** 2) + (self.y ** 2)) ** 0.5
    
    def __str__(self):
        return f'({self.x}, {self.y})'
    
    def halfway(self, target):
        mx = (self.x + target.x) / 2
        my = (self.y + target.y) / 2
        return Point(mx, my)
    
    def distanceFromPoint(self, target):
        return (((self.x - target.x) ** 2) + ((self.y - target.y) ** 2)) ** 0.5
    
    def reflect_x(self):
        return Point(self.x * -1, self.y)
    
    def slopeFromOrigin(self):
        return self.y/self.x if self.x != 0 else None
    
    def get_line_to(self, target):
        d = self.x - target.x
        if d == 0:
            return None
        m = (self.y - target.y)/d
        b = self.y - m * self.x
        return (m, b)
    
    def move(self, dx, dy):
        self.x = self.x + dx
        self.y = self.y + dy
````

Dati tre punti sulla circonferenza di un cerchio, trovare il centro e il raggio del cerchio.

Preso da [http://www.ambrsoft.com/trigocalc/circle3d.htm]

L'equazione del cerchio è data da 
$$ Ax^2 + Ay^2 + Bx + Cy + D = 0 $$
Sostituendo con i tre punti dati otteniamo il seguente sistema di equazioni
$$
\begin{vmatrix}
x^2 + y^2 & x & y & 1\\
x_1^2 + y_1^2 & x_1 & y_1 & 1\\
x_2^2 + y_2^2 & x_2 & y_2 & 1\\
x_3^2 + y_3^2 & x_3 & y_3 & 1
\end{vmatrix} = 0
$$

I coefficienti $A, B, C, D$ possono essere trovati risolvendo i seguenti determinanti
$$
A = \begin{vmatrix}
x_1 & y_1 & 1\\
x_2 & y_2 & 1\\
x_3 & y_3 & 1
\end{vmatrix}
\,
B = -\begin{vmatrix}
x_1^2 + y_1^2 & y_1 & 1\\
x_2^2 + y_2^2 & y_2 & 1\\
x_3^2 + y_3^2 & y_3 & 1
\end{vmatrix}
\,
C = \begin{vmatrix}
x_1^2 + y_1^2 & x_1 & 1\\
x_2^2 + y_2^2 & x_2 & 1\\
x_3^2 + y_3^2 & x_3 & 1
\end{vmatrix}
\,
D = -\begin{vmatrix}
x_1^2 + y_1^2 & x_1 & y_1\\
x_2^2 + y_2^2 & x_2 & y_2\\
x_3^2 + y_3^2 & x_3 & y_3
\end{vmatrix}
$$
I valori di $A, B, C, D$ sono calcolati tramite i determinanti
$$
A = x_1(y_2 - y_3) - y_1(x_2 - x_3) + x_2y_3 - x_3y_2 \\
B = (x_1^2 + y_1^2)(y_3 - y_2) + (x_2^2 + y_2^2)(y_1 - y_3) + (x_3^2 + y_3^2)(y_2 - y_1) \\
C = (x_1^2 + y_1^2)(x_2 - x_3) + (x_2^2 + y_2^2)(x_3 - x_1) + (x_3^2 + y_3^2)(x_1 - x_2) \\
D = (x_1^2 + y_1^2)(x_3y_2 - x_2y_3) + (x_2^2 + y_2^2)(x_1y_3 - x_3y_1) + (x_3^2 + y_3^2)(x_2y_1 - x_1y_2)
$$
Se $A \neq 0$ il centro del cerchio è dato da
$$ 
x = -\frac{B}{2A} \\
y = -\frac{C}{2A} 
$$
con raggio 
$$
r = \sqrt{\frac{B^2 + C^2 - 4AD}{4A^2}}
$$

````python
def circle(p1, p2, p3):
    A = p1.x * (p2.y - p3.y) - p1.y * (p2.x - p3.x) + p2.x * p3.y - p3.x * p2.y
    B = (p1.x**2 + p1.y**2) * (p3.y - p2.y) + (p2.x**2 + p2.y**2) * (p1.y - p3.y) + (p3.x**2 + p3.y**2) * (p2.y - p1.y)
    C = (p1.x**2 + p1.y**2) * (p2.x - p3.x) + (p2.x**2 + p2.y**2) * (p3.x - p1.x) + (p3.x**2 + p3.y**2) * (p1.x - p2.x)
    D = (p1.x**2 + p1.y**2) * (p3.x * p2.y - p2.x * p3.y) + (p2.x**2 + p2.y**2) * (p1.x * p3.y - p3.x*p1.y) + (p3.x**2 + p3.y**2) * (p2.x * p1.y - p1.x * p2.y)
    return (-B/(2*A), -C/(2*A), ((B**2 + C**2 - 4*A*D)/(4*A**2))**0.5) if A != 0 else None
````

````python
p1 = Point(32000, 2)
p2 = Point(6, 20)
p3 = Point(1500, 3)
(x, y, r) = circle(p1, p2, p3)
print (f"Dati i tre punti {p1}, {p2}, {p3}, l'equazione del cerchio passante per tali punti è")
print (f"(x - {x})^2 + (y - {y})^2 = {r}^2")
````
